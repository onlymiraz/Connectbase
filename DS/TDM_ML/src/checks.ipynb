```python
# function to see if primary carrier name is the same as previous month for each clean ID
def check_carrier_previous_month(cabs):
    """Create new column to denote if PRIMARY_CARRIER_NAME is the same in previous month.
    If yes, value is 1.
    If no, value is 0."""
    cabs["SAME_CARRIER_PREV_MONTH"] = cabs["PRIMARY_CARRIER_NAME"] == cabs.groupby("CLEAN_ID")["PRIMARY_CARRIER_NAME"].shift(1)
    cabs["SAME_CARRIER_PREV_MONTH"] = cabs["SAME_CARRIER_PREV_MONTH"].astype(int)

    # if row has first instance of a clean ID, set SAME_CARRIER_PREV_MONTH to 1
    cabs.loc[cabs.groupby("CLEAN_ID")["PRIMARY_CARRIER_NAME"].shift(1).isnull(), "SAME_CARRIER_PREV_MONTH"] = 1

    return cabs

df = check_carrier_previous_month(df)
```
